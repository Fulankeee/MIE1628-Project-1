```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Load data points from file
data_points = np.loadtxt(r"C:\Frank\UoT 2024-2025\MIE1628\A1\data_points.txt", delimiter=',')

# Cluster centers
cluster_centers = [
    (6.019058035946999, 7.40426255914861),
    (34.20406728296519, -1.3195706388069426),
    (-2.14542763330194, 9.171529421771337),
    (2.01077622673039, 12.790158977843065),
    (0.23804127333206698, 0.43573776470871767),
    (4.584713836139658, 5.129559701482909),
    (32.258840315815426, 3.9839980398093626),
    (31.685131777537812, 22.079927979273855),
    (8.895878957372847, 7.539907705049268)
]

# Use KMeans to assign each data point to the closest cluster center
kmeans = KMeans(n_clusters=len(cluster_centers), init=np.array(cluster_centers), n_init=1, max_iter=1)
kmeans.fit(data_points)
labels = kmeans.labels_

# Assign colors to clusters
cluster_colors = ['red', 'green', 'blue', 'orange', 'purple', 'brown', 'pink', 'cyan', 'yellow']
data_colors = [cluster_colors[label] for label in labels]

# Extract x and y coordinates for cluster centers
x_centers, y_centers = zip(*cluster_centers)

# Extract x and y coordinates for data points
x_data, y_data = data_points[:, 0], data_points[:, 1]

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, color=data_colors, marker='.', alpha=0.5, label='Data Points')

# Plot each cluster center with a different color
for i, (x, y) in enumerate(cluster_centers):
    plt.scatter(x, y, color=cluster_colors[i], marker='o', s=100, edgecolor='black', label=f'Cluster {i+1}')

# Labels and title
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Scatter Plot of Data Points and 9-Mean Cluster Centers with Cluster Assignments')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
